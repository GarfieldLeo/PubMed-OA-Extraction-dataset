```python
from nltk.stem.porter import PorterStemmer
import re
import json
import numpy as np
import matplotlib.pyplot as plt

f = open('../test.json','r')
abstracts = json.load(f)
f.close()
```

```python
# matching with word stemming
stemmer = PorterStemmer()
def match(extracted, ground):
    extracted_texts = set()
    for ph in extracted:
        phr=''
        words=ph.lower().split()
        for tkn in words:
            stmd=stemmer.stem(tkn)
            phr=phr+" "+stmd 
        if type(phr)==str:
            try:
                phr=phr.decode('utf-8')
            except:
                pass
        if bool(re.compile('^[a-zA-Z0-9_\-\s*]+$').match(phr)):    
            extracted_texts.add(phr.strip())

    ground_texts = set()
    for ph in ground:
        phr=''
        words=ph.lower().split()
        for tkn in words:
            stmd=stemmer.stem(tkn)
            phr=phr+" "+stmd 
        if type(phr)==str:
            try:
                phr=phr.decode('utf-8')
            except:
                pass
        if bool(re.compile('^[a-zA-Z0-9_\-\s*]+$').match(phr)):    
            ground_texts.add(phr.strip())        
    return len(extracted_texts.intersection(ground_texts))
```

```python
# iterate extracted kwds to compute score for each extraction.
#
actual = [0,0,0]
extracted = [0,0,0]
matched = [0,0,0]
f = open('yakekwds.json', 'r')
kwds = json.load(f)
f.close()
for index, ex in kwds.items():
    kwds_in = abstracts[index]["keywords_in"]
    index = 0
    for _, extracted_kwds in ex.items():
        score = match(extracted_kwds, kwds_in)
        actual[index] += len(kwds_in)
        extracted[index] += len(extracted_kwds)
        matched[index] += score
        index += 1
precision = np.divide(matched, extracted)
recall = np.divide(matched, actual)
f1 = 2 * np.divide((np.multiply(precision, recall)), precision + recall)

f = open('post_processed_score.txt', 'a')
f.write(f"Yake precision:{precision}\n")
f.write(f"Yake recall:{recall}\n")
f.write(f"Yake f1:{f1}\n")
f.close()
```
