```python
from nltk.stem.porter import PorterStemmer
import re
import json
import numpy as np
import matplotlib.pyplot as plt

f = open('../test.json','r')
abstracts = json.load(f)
f.close()
```

```python
# partial matching
stemmer = PorterStemmer()
def match(extracted, ground):
    extracted_texts = []
    for ph in extracted:
        phr=''
        words=ph.lower().split()
        for tkn in words:
            stmd=stemmer.stem(tkn)
            phr=phr+" "+stmd 
        if type(phr)==str:
            try:
                phr=phr.decode('utf-8')
            except:
                pass
        if bool(re.compile('^[a-zA-Z0-9_\-\s*]+$').match(phr)):    
            extracted_texts.append(phr.strip())

    ground_texts = []
    for ph in ground:
        phr=''
        words=ph.lower().split()
        for tkn in words:
            stmd=stemmer.stem(tkn)
            phr=phr+" "+stmd 
        if type(phr)==str:
            try:
                phr=phr.decode('utf-8')
            except:
                pass
        if bool(re.compile('^[a-zA-Z0-9_\-\s*]+$').match(phr)):    
            ground_texts.append(phr.strip())  
            
    score = 0
    for ex in ground_texts:
        tempList = [0]
        # split the extracted kwds into pieces
        tempC = list(re.split(' ', ex))        
        # for each word, loop through the groundtruth list to calculate score
        for act in extracted_texts:
            # split ground truth into pieces
            cc = list(re.split(' ', act))
            matched = len([w for w in cc if w in tempC])
            ss = 2*matched/(len(tempC)+len(cc))
            tempList.append(ss)
        ss = max(tempList)
        score += ss
    return score
```

```python
# iterate through the extracted kwds list and compute score for each extraction
# 
actual = [0,0,0]
extracted = [0,0,0]
matched = [0,0,0]

f = open('yakekwds.json', 'r')
kwds = json.load(f)
f.close()
for index, ex in kwds.items():
    kwds_in = abstracts[index]["keywords_in"]
    if kwds_in == []:
        continue
    
    index = 0
    # compute score for 5, 10, 15 kwds 
    for _, extracted_kwds in ex.items():
        score = match(extracted_kwds, kwds_in)
        actual[index] += len(kwds_in)
        extracted[index] += len(extracted_kwds)
        matched[index] += score
        index += 1
print(actual)
print(extracted)
print(matched)
precision = np.divide(matched, extracted)
recall = np.divide(matched, actual)
f1 = 2 * np.divide((np.multiply(precision, recall)), precision + recall)

f = open('post_processed_score_partial.txt', 'a')
f.write(f"Yake precision:{precision}\n")
f.write(f"Yake recall:{recall}\n")
f.write(f"Yake f1:{f1}\n")
f.close()
```
